**Answer 1** 

```python
#1. Can you find that there has been an improvement in mpg over the years ?
data1 = data.groupby('year')['mpg'].mean()
data2 = data1.to_frame()
data2.reset_index(drop = False, inplace = True)
data2
```

```python
plt.figure(figsize=(8,7))
sns.barplot(x=data['year'],y=data['mpg'],data=data)
#From the plot below we can conclude that there has been an improvement in the mpg over the yers.
```

**Answer 2**

```python
#2. Does cubic inches of a vehicle affect the vehicle's time to reach 60 Miles per hour?
data3 = data.groupby('time-to-60')['cubicinches'].mean()
data4 = data3.to_frame()
data4.reset_index(drop = False, inplace = True)
data4
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x=data['time-to-60'],y=data['cubicinches'],data=data)
#From the plot we can infer that the vehicle's time to reach 60 miles per hour is affected by the cubic inches
```

**Answer 3**

```python
#3. Does a specific country prefer vehicles with more horsepower?
data5 = data.groupby('brand')['hp'].mean()
data6 = data5.to_frame()
data6.reset_index(drop = False, inplace = True)
data6
```

```python
plt.figure(figsize=(5,7))
sns.barplot(x=data['brand'],y=data['hp'],data=data)
#The plot clearly shows that a specific country does prefer vehicles with a particular range of horsepower.
```

**Answer 4**

```python
#4. How does horsepower affect the average of the vehicle ?
data7 = data.groupby('hp')['mpg'].mean()
data8 = data7.to_frame()
data8.reset_index(drop = False, inplace = True)
data8
```

```python
plt.figure(figsize=(35,15))
sns.lineplot(x=data['hp'],y=data['mpg'],data=data)
```

```python
plt.figure(figsize=(35,15))
sns.barplot(x=data['hp'],y=data['mpg'],data=data)
#From the plot below we infer that with increase in horsepower the average of the vehicle mostly decreases
```

**Answer 5**

```python
#5. How does the cubic inches affect the average of the vehicle ?
data9 = data.groupby('cubicinches')['mpg'].mean()
data10 = data9.to_frame()
data10.reset_index(drop = False, inplace = True)
data10
```

```python
plt.figure(figsize=(35,15))
sns.lineplot(x=data['cubicinches'],y=data['mpg'],data=data)
```

```python
plt.figure(figsize=(45,15))
sns.barplot(x=data['cubicinches'],y=data['mpg'],data=data)
#the effect is shown in the graph below
```

**Answer 6**

```python
#6. What is the effect of the number of cylinders on the mpg?
data11 = data.groupby('cylinders')['mpg'].mean()
data12 = data11.to_frame()
data12.reset_index(drop = False, inplace = True)
data12
```

```python
sns.lineplot(x=data['cylinders'],y=data['mpg'],data=data)
```

```python
plt.figure(figsize=(4,5))
sns.barplot(x=data['cylinders'],y=data['mpg'],data=data)
#With increase in the number of cylinders from 3 to 4 the mpg decreases but on increasing more number of cylinders the mpg starts decreasing
```

**Answer 7**

```python
#7.What relation between cubic inches, number of cylinders and horsepower can you observe from the data?
aggregation_functions ={ 'cylinders' :'mean', 'hp': 'mean'}
data13=data.groupby('cubicinches').aggregate(aggregation_functions)
data13.reset_index(drop = False, inplace = True)
data13
```

```python
plt.scatter(x=data['cubicinches'],y=data['cylinders'])
#x-axis shows the cubicinches and y-axis shows the number of cylinders
```

```python
plt.scatter(data['hp'],data['cylinders'])
#x-axis shows the horsepower and y-axis shows the number of cylinders
```

```python
#From the above scatter plots it is clear that horsepower is directly proportional to cubic inches and number of cylinders
```

**Answer 8**

```python
#8.How much proportion of the total records is occupied by vehicles with 8 cylinders?
z = data['cylinders'].value_counts()
z1=z.to_frame()
z1.reset_index(drop = False, inplace = True)
z1
```

```python
plt.title("Percentage of records occupied by vehicles with cylinders", fontsize=14);
plt.pie(z1['cylinders'], explode=[0.05,0.05,0.05,1.5,1.5], autopct='%0.1f', labels=z1['index'])
plt.show()
#29.1% of the total is occupied by vehicles with 8 cylinders
```
